```python
import pandas as pd

data=pd.read_csv("../data/raw/aerostream.csv")
```

```python
data.isna().sum()
```

```python
import re 

def clean_text(text):
    text=text.lower()
    text=re.sub(r"http\S+|www\S+","",text)
    text = re.sub(r"@\w+", "", text)                
    text = re.sub(r"#\w+", "", text)                
    text = re.sub(r"[^a-zA-Z\s]", "", text)         
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
data["clean_text"]=data["text"].apply(clean_text)
```
